Cargamos inicialmente las libreriras a utilizar

```
import numpy as np
from scipy import signal
```

Definimos el sistema linealizado alrededor del punto de equilibrio

```
A = [[-2.0,-6.0],[-8.0,-8.0]]
print(np.linalg.eig(A)[0])
B = [[8.0],[0.0]]
C = [[0.0,1.0],[1.0,0.0]]
D = [[0.0],[0.0]]
sys = signal.StateSpace(A,B,C,D)
t,y = signal.step(sys)
```

```
import matplotlib.pyplot as plt
plt.figure(1)
plt.plot(t,y[:,0]+2.0,'r--',linewidth=3)
plt.plot(t,y[:,1]+2.0,'b-',linewidth=3)
plt.legend(['y1','y2'],loc='best')
plt.ylabel('Ejercicio Espacio de Estados Linealizado')
plt.xlabel('Tempo')
plt.show()
```
